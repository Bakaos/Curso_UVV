# **3) Gerar X e y: X com as features dependentes, y com o target**

```python
y = df['X5']
x = pd.DataFrame()
x = df.drop(['X5'],axis=1)
x.head()
```

# **4) Dividir o dataset em treino e teste (30%), com random_state=42 X_train, X_test, y_train, y_test**

```python
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.3,random_state=42)
```

# **5) Criar um modelo de regressão linear usando sklearn**

```python
lm = LinearRegression()
```

# **6) Treinar o modelo com os dados de treino**

```python
lm.fit(x_train, y_train)
```

# **7) Realizar a predição com os dados de teste**

```python
# Para validar o modelo no conjunto de teste, você precisa utilizar o método predict()
y_pred_test = lm.predict(x_test)
```

# **8) Crie um diagrama de dispersão (scatterplot) dos valores reais de teste em relação aos valores de teste preditos.**

```python
# Criar um diagrama de dispersão (scatterplot) dos valores reais de teste em relação aos valores preditos.
plt.scatter(y_test, y_pred_test)
plt.xlabel('y real')
plt.ylabel('y predito')
```

```python
# Criar um diagrama de dispersão (scatterplot) dos valores reais de teste em relação aos valores preditos.
# Foi adicionado a Linha de Equivalência para facilitar a visualização da dispersão dos dados
start_point = max(min(y_test), min(y_pred_test))
final_point = min(max(y_test), max(y_pred_test))
a = np.linspace(start_point, final_point, 20)
b = a
plt.plot(a, b, color='red', label='Line of Equality (y=x)')
plt.scatter(y_test, y_pred_test)
plt.xlabel('y real')
plt.ylabel('y predito')
plt.show()
```

# **9) Avalie o modelo com MAE, MSE, RMSE, R2 ajustado, MAPE para os dados de teste. O que se pode concluir?**

```python
#Avaliar o Modelo, os valores maximos e minimos foram adicionados para referencia
print('Valores Minimo e Maximo previstos:\nmin: ',min(y_pred_test) ,'\nMax: ', max(y_pred_test))
print('Valores Minimo e Maximo esperados:\nmin: ',min(y_test) ,'\nMax: ', max(y_test))
print('MAE:', metrics.mean_absolute_error(y_test, y_pred_test))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_test))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)))
print('R:', metrics.r2_score(y_test,y_pred_test))
```
