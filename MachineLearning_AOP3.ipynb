```python
target = df['deposit_cat']
dt = pd.DataFrame()
dt = df.drop(['deposit_cat'],axis=1)
x_train, x_test, y_train, y_test = train_test_split(dt,target,test_size=0.5)
print('\"x\" Shape:\nTrain:',x_train.shape,'\t Test: ', x_test.shape,'\n\"y\" Shape:\nTrain: ',y_train.shape,'\t Test: ', y_test.shape)
```

## 1.2 - Treine o modelo baseado em arvore de decisão com seus parâmetros padrões.

```python
clf = tree.DecisionTreeClassifier(random_state=1)
clf = clf.fit(x_train,y_train)
for feature,importancia in zip(x_train.columns,clf.feature_importances_):
    print("{}:\t{}".format(feature, importancia))
# Parâmetros do  classificador DecisionTreeClassifier
# criterion: medida de qualidade da divisão
# splitter: estratégia utilizada para dividir o nó de decisão
# max_depth: profundidade máxima da árvore
# min_samples_split: número de amostras mínimas para considerar um nó para divisão
# min_samples_leaf: número de amostras mínimas no nível folha
```

```python
# Caracteristicas da Árvore

n_nodes = clf.tree_.node_count
children_left = clf.tree_.children_left
children_right = clf.tree_.children_right
threshold = clf.tree_.threshold

print('Numero de nodos: ',n_nodes,'\n')
print('Nodos a esquerda: ',children_left,'\n')
print('Nodos a direita',children_right,'\n')
print('threshold',threshold,'\n')
```

## 1.3 - Computar a acurária dos dados de treino e de teste

```python
y_pred_test = clf.predict(x_test)
y_pred_train = clf.predict(x_train)
y_pred_test
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
print('Acuracia do teste:\t\t',accuracy,'\n')

accuracy = accuracy_score(y_train, y_pred_train)
print('Acuracia do treinamento:\t',accuracy,'\n')
```

```python
#Matriz de confusão,indicando valor absoluto de erros e acertos, a diagonal de valores corretos mostram as previsões feitas corretamente e a diagonal de valores falso os erros.

conf_matrix_train = confusion_matrix(y_train, y_pred_train)
conf_matrix_test = confusion_matrix(y_test, y_pred_test)
print('modelo da matriz de confusao:\n\t\tPredito Negativo\tPredito Positivo\nReal Negativo\tNegativo Correto\tFalso Positivo\nReal Positivo\tFalso Negativo\t\tPositivo correto')
print('\nmatriz de erro do treino:\n',conf_matrix_train)
print('\nmatriz de erro do teste:\n',conf_matrix_test)
```

```python
# Outras métricas:
accuracy = accuracy_score(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)

print('Acuracia: ',accuracy,'\n')
print('Precisao: ',precision,'\n')
print('Sensibilidade (positivo): ',recall,'\n')
print('F1 Score: ',f1,'\n')
```

Durante a fase de treinamento o algoritimo acertou 100% dos testes, durante a fase de teste ele teve uma performance de 74,05%, indicando overfitting

## 3.3 - Utilize a função compara_modelos() conforme o código abaixo, bem como os prints para a chamada desta função com diferentes valores de profundidade.

```python
def compara_modelos(maxdepth,x_train, x_test, y_train, y_test):
  if maxdepth == 0:
    dt = tree.DecisionTreeClassifier(random_state=1)
    dt.fit(x_train, y_train)
    train_score = dt.score(x_train, y_train)
    test_score = dt.score(x_test, y_test)
  else:
    dt = tree.DecisionTreeClassifier(random_state=1, max_depth=maxdepth)
    dt.fit(x_train, y_train)
    train_score = dt.score(x_train, y_train)
    test_score = dt.score(x_test, y_test)
  return train_score,test_score
```

## 3.4 - Faça as alterações necessárias para utilizar os dados divididos anteriormente.

```python
for i in range(10):
  treino, teste = compara_modelos(i,x_train, x_test, y_train, y_test)
  print('Train, test scores' ,i,':\t',treino, teste)
  print('Diferença:\t', (treino-teste),'\n')
```
